```python
# 
# 1. load
# 
import pandas as pd
galaxy = pd.read_csv('https://web.stanford.edu/~hastie/ElemStatLearn/datasets/galaxy.data')
#vs
#iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/bezdekIris.data')

print('head',galaxy.head())


```

```python
# 
# metadata
# 

print('type',type(galaxy))
print('info\n',galaxy.info)
print('keys',galaxy.keys())
#vs
print('columns',galaxy.columns)
```

```python
# 
# may analize each feature
# 
print('mean',galaxy["radial.position"].mean())
print('value_counts\n',galaxy["radial.position"].value_counts().head())
print('shape',galaxy["radial.position"].value_counts().shape)
```

```python
# 
# descriptive stats
# 

print('describe',galaxy.describe())
```

```python
# 
# vs
# 

print('forEach feature')
print(' SUM\n',galaxy.sum())
print(' mean\n',galaxy.mean())
print(' min\n',galaxy.min())
print(' max\n',galaxy.max())
```

```python
# 
# excerpts 
#

print('cumsum\n',galaxy["radial.position"].cumsum().head())
#
print('diff\n',galaxy["radial.position"].diff().tail())
```
